# Homomorphic Functions

## Define

Let $\sum_1$ and $\sum_2$ be 2 alphabets. A homomorphism from $\sum_1$ to $\sum_2$ is a function

$h : \sum^*_1 \implies \sum^*_2$ $s.t.$

1. $h(\lambda)$
2. $h(w_1 \cdot w_2) = h(w_1) \cdot h(w_2)$ $\forall w_1, w_2 \in \sum^*_1$
